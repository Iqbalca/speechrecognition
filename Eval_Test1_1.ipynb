```python
from huggingsound import SpeechRecognitionModel
```

```python
model = SpeechRecognitionModel("jonatasgrosman/wav2vec2-large-xlsr-53-german")
```

```python
references = [
    {"path": "AG_dialect.wav", "transcription": "ich möchte noch darauf hinweisen dass es unterschiede in der aussprache oder unterschiedliche wörter innerhalb ein und derselben dialektgruppe gibt zum beispiel klingt eine person aus tirol in der aussprache ganz anders als jemand aus der steiermark obwohl beide bundesländer zur südbayerischen dialektgruppe gehören des des weiteren gibt es unterschiede innerhalb ein und desselben bundeslandes die aussprache"}
]    

evaluation = model.evaluate(references)

```
